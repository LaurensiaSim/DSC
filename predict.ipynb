```python
import pandas as pd
import xgboost as xgb
```

```python
loaded_model = xgb.XGBClassifier()
loaded_model.load_model("xgboost_model.json")
loaded_model
```

```python
data=pd.read_csv('./data/test.csv',header=0)
data
```

```python
selected=['id','DC024', 'DC025', 'DC205', 'DC220', 'DC226', 'DC230a', 'DC232b',
       'DC235', 'DC237', 'DC237a', 'DC237b', 'DC237c', 'DC237d', 'DC237f',
       'DC241', 'DC244', 'DC246', 'DC252', 'DC109']
test=data[selected]
test
```

```python
test.iloc[:,1:]
```

```python
predict=loaded_model.predict(test.iloc[:,1:])
predict
```

```python
submission=pd.read_csv('./data/sample_submission.csv')
submission['DC201']=predict
submission
```

```python
submission['DC201'].value_counts()
```

```python
submission['DC201']=submission['DC201'].apply(lambda x:'Layak Minum' if x==0 else 'Tidak Layak Minum')
submission
```
